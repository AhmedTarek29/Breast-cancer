```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('breast-cancer.csv')
df.head() #out first 5 rows of dataframe
```

```python
#Shape of the Dataset
df.shape
```

```python
#Describtion of data
df.describe()
#for each column:count(number of not-empty values) ,mean(avg value) ,std(standard deviation) ,min(min value) ,25%(25%percentile*) ,50% (50% percentile*) ,75%(75% percentile*) ,max(max value)
```

```python
#set the ID column to be the index of the dataframe
df = df.set_index('id')
```

```python
df.drop(['Unnamed: 32'],axis=1,inplace= True)
#to drop specified labels from rows or columns
```

```python
# Columns in the dataset
df.columns
```

```python
#Count the diagnosis column such that M = 1, B = 0
df['diagnosis'] = df['diagnosis'].apply(lambda x: '1' if x == 'M' else '0')
df['diagnosis'] = df['diagnosis'].astype("float64")
```

```python
#The number of Benign and Maglinant cases from the dataset.
print(df.groupby('diagnosis').size())
```

```python
# Count of null value per column
missing_values_count = df.isnull().sum()
missing_values_count[0:31]
```
